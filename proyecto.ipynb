```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("syedmtalhahasan/global-urban-air-quality-index-dataset-2015-2025")

print("Path to dataset files:", path)
```

```python
import pandas as pd
df=pd.read_csv(path+"/global_air_quality_dataset.csv")
df.head()
```

```python
# prompt: categorizar en una columna calidad del aire teniendo en cuenta que si PM2.5 es menor a 50 bueno, entre 51 y 100 es moderado, entre 101 y 150 es no sano para grupos sensibles, entre 151 y 200 es no sano, entre 201 y 300 es muy insano, y mas de 301 es peligroso y muestrame toda la tabla

# Assuming 'df' is your DataFrame with the air quality data and a column named 'PM2.5'
def categorize_air_quality(row):
  """Categorizes air quality based on PM2.5 levels."""
  pm25 = row['PM2.5 (µg/m³)']
  if pm25 < 50:
    return 'Bueno'
  elif 51 <= pm25 <= 100:
    return 'Moderado'
  elif 101 <= pm25 <= 150:
    return 'No sano para grupos sensibles'
  elif 151 <= pm25 <= 200:
    return 'No sano'
  elif 201 <= pm25 <= 300:
    return 'Muy insano'
  else:
    return 'Peligroso'


# Apply the function to create a new column 'Calidad del Aire'
df['Calidad del Aire'] = df.apply(categorize_air_quality, axis=1)

# Show the entire table with the new column
df
```

```python

df = df.drop(['Date', 'City', 'Country', 'PM2.5 (µg/m³)'], axis=1)

# Show the entire table with the new column and removed columns
df
```
